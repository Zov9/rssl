```python
# Define the path to your data file
file_path = '/data/lipeng/ABC/txt/cf100_0501_1623t3.txt'

# Initialize a list to hold the class indexes for each epoch
worst_classes_per_epoch = []
worst_k_predicted_per_epoch = []
# Read the text file
with open(file_path, 'r') as file:
    lines = file.readlines()

# Process the lines
i = 0
while i < len(lines):
    line = lines[i].strip()
    # Check if the line contains the start of the recall section
    if line.startswith('Worst k predicted'):
        numbers_line = lines[i + 1].strip()
        # Extract the numbers within the parentheses
        numbers = numbers_line.strip('()').split(', ')
        # Convert the extracted strings to integers
        worst_k = [int(num) for num in numbers]
        worst_k_predicted_per_epoch.append(worst_k)
        i += 100

    elif line.startswith('5 Classes with Smallest Recall'):
        # Skip the title line and extract the next 20 lines for class indexes
        class_indexes = []
        for j in range(1, 21):  # Assuming there are always 20 lines following the title
            class_info = lines[i + j].strip()
            class_index = int(class_info.split()[1])  # Extract the class index
            class_indexes.append(class_index)
        worst_classes_per_epoch.append(class_indexes)
        i += 1100  # Move to the next section
    
    else:
        i += 1  # Continue to the next line

# Now, worst_classes_per_epoch contains the class indexes for all epochs
# You can process this list further as needed, or print it out
for epoch, classes in enumerate(worst_classes_per_epoch, start=1):
    print(f"Epoch {epoch}: Worst classes indexes - {classes}")
for epoch, k_predicted in enumerate(worst_k_predicted_per_epoch, start=1):
    print(f"Epoch {epoch}: Worst k predicted - {k_predicted}")
```

```python
def cal_sim(rw1,pw1):
    sim1 = []
    for i, j in zip(rw1, pw1):
        simt = len(set(i).intersection(set(j))) / len(i)
        sim1.append(simt)
    return sim1
```

```python
sim1 = cal_sim(rw1,pw1)
sim2 = cal_sim(rw2,pw2)
sim3 = cal_sim(rw3,pw3)
sim4 = cal_sim(rw4,pw4)
sim5 = cal_sim(rw5,pw5)
sim6 = cal_sim(rw6,pw6)
sim7 = cal_sim(rw7,pw7)
```
